#**Introduction to Chatbot & Dialogflow**
----

## What is our **goal** ?

Our goal is to create a chatbot using DialogFlow.

## What is a **chatbot** ?

A chatbot is a software able to **conduct a conversation** via an interface (either textual or oral). They are designed to **simulate human behaviour** in the setup of a conversation. They are mostly used for customer services or information sharing. They can work either with sophisticated language systems or via keywords and patterns from a database.

## What is **Dialogflow** ?

DialogFlow is a **Google cloud** service used to build **conversational interfaces**
such as **chatbots** for websites, mobile applications, IoT devices, and messaging platforms. Dialogflow is an end-to-end tool for building conversational experiences powered by natural language understanding. By using Google’s machine learning, Dialogflow helps us transforming rich and natural human language into machine-readable data.

For example, booking an appointment with the doctor can be done instantly via a personalized, cost-efficient service. By building a conversational experience powered by AI, patients can interact and chat directly with the doctor’s chatbot to book an appointment, ask questions or make small talk. The main thing that has to be done to develop the chatbot is to provide a dozen patients dialogues examples. Then, machine learning is used to understand what patients want and to respond in the most efficient way. Finally, the doctor can test, launch the chatbot on his website and track analytics of user experience.

## What are the **fundamental concepts** ?

The ideal interface should be able to **interpret and process the natural language** of end-users to infer what they want.

# **Components of a Dialogflow's Chatbot**
----

## **Agents**
In Dialogflow, a virtual agent is a natural language understanding module
handling conversation with end-users. In order to transform end-user text to
structured data, an agent needs to be designed and trained according to the
service that wants to be delivered.

In dialog flow, an agent is what represents an entire chatbot application. For example, Siri is like an agent.
## **Intents**

The reaction of a bot is determined by the intents you give him. An intent
represents one way the bot will react to specific user’s intention. You have to define many intents for a single bot. In each of these, you define the way the bot react to a given set of end-user expression. And given the expression input by the user, DialogFlow will match it to the best intent you defined in your agent. So the capacity of a bot to handle a conversation depends on the intents you gave him.

For example, the cocktail agent we have created recognizes and responds to
end-user questions about cocktail recipes. If an end-user asks for the recipe of a Mojito cocktail, Dialogflow will do the matching between the end-user expression and the corresponding intent. Intents can extract critical data from the end-user expression like ingredients for the desired cocktail recipe.

An intent has :

-   **Training phrases**: These are sentences that the user might use in order to get some information. End-user expression that is similar to these training phrases let Dialogflow match the end-user expression to the most likely *intent* depending on the training phrases given and the ones added by Dialogflow built-in machine learning.

-   **Action**: An action is possibly defined for each *intent* to trigger specific actions or events in the system.

-   **Parameters**: From the end-user expression, some critical values considered as structured data are extracted based on the defined *entity type* as parameters to generate responses.

-   **Responses**: These are responses to the training phrases and the end-user expression that are matched with this particular intent. Responses could be text, speech, photos or videos.

Intent matching helps us recognizing what the users want. By creating an intent, we define what the user might request. For example, he might want to check prices, book an appointment, check medical records, pay a bill, ask for a recipe, cancel an appointment, or get office hours. For each intent related to a user request, a few examples of various sentences the user might use to communicate the same request has to be defined.

In order to recognize what a user wants, whenever he says something, our agent will match his expression to the most suitable intent according to the example of sentences given (training phrases). Then, once the most-suitable intent is found, static or dynamic responses can be provided to the user.

A doctor’s office agent can have a few or a thousand intents, but each of them should recognize a specific user need. The more used your chatbot is the smarter it becomes because more training phrases are added and false intent classification can be corrected.

### **Follow-up intents**

These are used to associate a child’s intent to its parent intent and
automatically set input (to child intent) /output (to parent intent) contexts.
Typically, predefined follow-up intents are “yes/no” answers. Matching a child’s intent is possible if the parent’s intent was matched with the previous end-user expression.

### **Fallback intents**

These are triggered if an end-user expression doesn’t match any available
intent. A common practice is to use these fallback intents to guide the user’s in the right conversational direction.
## **Entities**

Entity extraction helps us identify detailed pieces of information and useful
parameters mentioned by users. When a user expresses his need, he wants
something specific. This is why just matching an intent might not be enough to give the user what he’s asking for. Extracting important details from a user’s statement helps us figure out more precisely the user’s need or intent. In Dialogflow, some built-in system entity types extract automatically dates, times, place names or numbers, but we can easily add custom entities into our agent by providing a list of words corresponding to the entity type.

While adding possible users’ expression to our intent, Dialogflow should
recognize entities present in order to extract not only the entity but the value of the matching entity and possibly use it in the backend code.

In order to know what to do to give the users response to an intent, setting
actions and parameters is primordial to collect and store variables that we
need. So, we can either use those variables to provide a static answer or to use it back-end code for providing dynamic answers.

For example, we defined a “name” entity that can match the cocktail’s name with our cocktail agent.

Here is a non-exhaustive list of built-in **system entities** are:

-   Date and time \@sys.time \@sys.date

-   Numbers \@sys.numbers \@sys.flight-number

-   Amounts with units \@sys.unit-currency \@sys.unit-length

-   Geography \@sys.address

-   Contacts \@sys.email \@sys.phone-number

-   Names \@sys.given-name \@sys.last-name



**Developer entities** can be customized in order to provide the system through a CSV file or API with a list of words or values that can be used for backend code. Synonyms of values can be provided manually or automatically by enabling the automated expansion that lets the agent capture words beyond those defined in the entities’ lists.

User entities can also be helpful to define a specific user session like
previous orders.

## **Contexts**

In order for Dialogflow to understand what an end-user is referring to and match the exact intent, contexts should be defined. Setting input and output contexts for an intent guides the flow of the conversation. Output contexts for an intent become active once this intent is matched with the end-user expression. Those active contexts can then be used in input contexts of other intents to increase the probability of matching. Contexts expire automatically after 20 minutes or a predefined number of conversational turns. They can be used to store data like values of parameters of an entity type.

By using context, it’s possible to store and access variables to exchange
information from one intent to another one.


## **Fulfilment for integrations**

By using fulfilment, dynamic responses can be provided to the end-user once the intent is matched instead of a defined static response. By enabling fulfilment for an intent, an external defined service can be the response to that intent. When this intent is matched, a request to the defined webhook service with specific information related to the matched intent is sent to perform any actions.

For example, if an end-user wants to have the cocktail recipe for a mojito, our service can check the database of the CocktailDB API and then send the recipe to the end-user.

# **Overall mechanisms**
----
##**General functioning**

![Texte alternatif…](https://drive.google.com/uc?id=1vwCeNZIc7A7z3s3rixD3fHLuI3nWPIWy)

1.  The end-user types “Give me the recipe for the cocktail Virgin Mojito”.

2.  Dialogflow matches this expression to the intent calle “Cocktail-specific” and extracts parameters like “virgin” for the entity called “Alcoholic”.

3.  Dialogflow sends a webhook request containing information about the “Cocktail-specific” intent, “Alcoholic” parameters and response“# predefined for this intent to our webhook service.

4.  Our webhook service performs an external API call.

5.  Our webhook service sends the webhook response containing the response to send to the end-user to Dialogflow

6.  Dialogflow sends the response to the end-user.

7.  The end-user sees the response.

Fulfilment is the code needed to interface with a back-end service to provide
dynamic answers to the end-users.

## **Overall coherence of the conversation**

Bringing everything together and shaping the flow of a conversation, make them appear rich and natural is more than a single statement or response. A conversation is a back and forth negotiation dialog.

We can distinguish two types of dialog:

-   *Linear* to collect all *specifically required information or set of facts for actions* like booking an appointment.

    -   Book an appointment with required entities and prompt questions (feature slot filling)

-   *Non-linear* to look like a real natural conversation with natural speaking context

    -   Interpretation of users’ expression in the context of what was said
        before.

## **Backend operations and integration**

By using values of parameters, it’s possible to take actions in the backend
(like booking an appointment in a Google calendar). Moreover, it’s possible to deliver a personalized response to the user’s as well programmatically by using fulfilment.

To do so, a web server exposing a single HTTP endpoint has to be created
(WEBHOOK). By providing this HTTP endpoint, Dialogflow knows the connection is required. The built-in editor is Firebase. In order to give a customized answer to a user’s query, little Javascripts have to be written and deployed into Cloud Functions for Firebase (a click). Then, fulfilment should be enabled for intents that need an external API call for example.

When giving each intent an action name, it’s easier for the fulfilment of which intent was triggered. Once one of these intents is matched, a JSON request will be sent to the WebHook containing the user’s expression, values of extracted entities and action name to recognize which intent was matched. The generated response will be sent to the user through Dialogflow. By using the WebHook, it’s also possible to set and remove contexts and parameter values to better control the conversational flow.

Finally, one-click integrations are possible with Facebook, Slack or Telegram as well as built-in analytics from those resources.

## **The Training**

Based on the training phrases you give him and the google machine learning AI a bot will be able to recognize and understand some patterns that will allow him to link more effectively user inputs to intents. But sometimes this will not be enough and the bot will link your inputs to the wrong intent or won’t be able to link it at all.  
Another way to help your bot learning is via the “Training” tab. In it, you will find a collection of all the inputs and discussion with the bot. By expanding one conversation you will see every inputs and within which entities did the bot recognize and to which intent he linked it. Then you can manually correct the intent classification or delete the end-user expression from the learning pool.

## **5. Integrate with an external API**

Our Dialogflow is setup but it is not yet linked to the API. That means that when someone asks for the recipe of a cocktail, our chatbot will return nothing. So, we need to
configure our chatbot so that it sends request to the API from the different
intents and then return the response to the user. 

The API we will use is the following one : https://www.thecocktaildb.com/api.php

Here’s the procedure to do so:

1.  Enable the “**webhook call for this intent** “ in the intent you want to communicate with an API.

  ![API](https://drive.google.com/uc?id=1vSVeC6aXGyuzRtbsK6crdXu-OMCEc0A4)

2.  Activate the *“Inline Editor”* in the **fulfilment** section. This will allow us to create our own functions. Make sure that you have upgraded **Firebase** to premium’s version. To access **Firebase**, click on *“View execution logs in the Firebase console”* below the **inline editor**

  ![API](https://drive.google.com/uc?id=1ukaopz__LIKMzULBUFGuANGIxom-_L-f)

3.  In *package.json*, add the axios dependency and then import it into index.js. The axios package makes it easy to send HTTPS requests to APIs.

  ![API](https://drive.google.com/uc?id=15mRE3C1Ce7moyF7a_HvaKAfyBEb1kfm6)
  Also add in *index.js* : 
  ```javascript
  const axios = require("axios");
  ```


4.  In *index.js* some functions have already been written. However, we will need to define a number of other functions in order to communicate with the API.

  >Function **WriteCocktail**

  This function takes as argument : the **agent** (allows us to interact with the sculpin) and the **drink** (the JSON object that will be returned by the API).
  The agent.add() function is used to respond in the chatbot.

  ```javascript
function writeCocktail(agent, drink) {  
   agent.add("The " + drink.strDrink);  
   agent.add(" ");  
   agent.add("Instructions");  
   agent.add("");  
   agent.add(drink.strInstructions);  
   agent.add(" ");  
   agent.add("Ingredients");  
   if (drink.strIngredient1 != null) {  
   agent.add(drink.strIngredient1);  
   }  
   if (drink.strIngredient2 != null) {  
   agent.add(drink.strIngredient2);  
   }  
   if (drink.strIngredient3 != null) {  
   agent.add(drink.strIngredient3);  
   }  
   if (drink.strIngredient4 != null) {  
   agent.add(drink.strIngredient4);  
   }  
   agent.add(new Image(drink.strDrinkThumb));  
   }
```

  >Function **GetCocktail**
  
  This function takes only the agent as argument. We first recuperate the parameter “cocktail_name” by using : *agent.parameters.NAME_OF_PARAMETER*.
  Then, we use the axios package to send an HTTP request of type GET to the API, which is **thecocktaildb** in our case.
  Finally we give our another function **writeCocktail** the drink which has been returned by the api.

  ```javascript
   function getCocktail(agent) {  
   agent.add("test");  
   console.log("function called");  
   const name = agent.parameters.cocktail_name;  
   if (name != "") {  
   return axios  
   .get( "https://www.thecocktaildb.com/api/json/v1/1/search.php?s=\${name}"  
   )
   .then(result => {  
   writeCocktail(agent, result.data.drinks[0]);  
   });  
   }  
   }
```
